<h3> ETL preparation pipeline </h3>

```python
import pandas as pd
import numpy as np
```

<h3>Extract</h3>

```python
msg = pd.read_csv("messages.csv")
msg.head()
```

```python
cat = pd.read_csv("categories.csv")
cat.head()
```

<h3>Transform Data</h3>

```python
df=pd.merge(msg,cat,on="id",how="outer")
df.head()
```

Separate Categories column by delimiter ;

```python
categories =df['categories'].str.split(';',expand=True)
categories.head()
```

```python
name = categories.head(1)
```

```python
names = name.applymap(lambda x: x[:-2]).iloc[0,:]
print(names)
```

Rename columns of categories dataframe

```python
categories.columns = names
categories.head()
```

```python
categories.related.value_counts()
```

Converting categories value to numeric value

```python
for column in categories:
    categories[column]=categories[column].astype(str).str[-1]
    categories[column]=categories[column].astype(int)
```

```python
categories.head()
```

```python
categories.related.value_counts()
```

Makiing the column value of 2 as 1 for related column

```python
categories["related"]=categories["related"].replace(to_replace=2, value =1)
```

```python
categories.related.value_counts()
```

Replacing category column in merged dataframe with new category columns

```python
df.drop('categories', axis = 1, inplace = True)
df.head()
```

Concatenating the new categories dataframe with df dataframe


```python
df = pd.concat([df,categories], axis = 1)
df.head()
```

Removing duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df1= df.drop(['id','message','original','genre'],axis=1)
```

```python
categories_counts = df1.sum(axis=0)
```

```python
categories_counts
```
